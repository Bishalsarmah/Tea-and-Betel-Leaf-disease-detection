```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
```

```python
# Path to your image
image_path = r"C:\Users\vivek\Desktop\Bacterial_Leaf_Spot_Disease(17).jpg"  # Replace with your image path

# Load and preprocess the image
img = load_img(image_path)
x = img_to_array(img)
x = np.expand_dims(x, axis=0)

# Define augmentations (width+height shift combined)
augmentations = {
    'Original Image': None,
    'Width Shift (20%)': ImageDataGenerator(width_shift_range=0.2, fill_mode="None"),
    'Height Shift (20%)': ImageDataGenerator(height_shift_range=0.2),
    'Rotation (30°)': ImageDataGenerator(rotation_range=50),
    'Brightness (0.5-1.5)': ImageDataGenerator(brightness_range=[0.5, 0.6]),
    'Zoom (15%)': ImageDataGenerator(zoom_range=0.3),
    'Horizontal Flip': ImageDataGenerator(horizontal_flip=True),
    'Vertical Flip': ImageDataGenerator(vertical_flip=True)
}

# Prepare subplot grid (2 rows, 4 cols)
num_aug = len(augmentations)
cols = 4
rows = int(np.ceil(num_aug / cols))

fig, axes = plt.subplots(rows, cols, figsize=(16, 8))
axes = axes.flatten()

# Apply augmentations and plot
for ax, (title, datagen) in zip(axes, augmentations.items()):
    if datagen is None:
        aug_img = x[0].astype('uint8')
    else:
        iter_img = datagen.flow(x, batch_size=1)
        aug_img = next(iter_img)[0].astype('uint8')
    ax.imshow(aug_img)
    ax.set_title(title)
    ax.axis('off')

# Hide any unused subplots
for ax in axes[num_aug:]:
    ax.axis('off')

# Tight layout with minimal gaps
plt.tight_layout(pad=0.5)
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```
