# Combining Tea and Betel Leaf Datasets

```python
import os
import shutil
import zipfile
from google.colab import drive
```

```python
# --- 1. Mount Google Drive ---
drive.mount('/content/drive')
```

```python
# --- 2. Define Paths ---
# *Update these paths to the actual locations of your ZIPPED datasets in Google Drive*
tea_zip_path = '/content/drive/MyDrive/My Datasets/Tea/Tea_Leaf_Final_Dataset.zip'      # Zipped Tea Leaf dataset
betel_zip_path = '/content/drive/MyDrive/My Datasets/Betel/Betel_Leaf_Final_Dataset (1).zip'  # Zipped Betel Leaf dataset
extracted_tea_folder = '/content/extracted_tea'
extracted_betel_folder = '/content/extracted_betel'

# --- Paths within the final COMBINED dataset directory ---
final_combined_dir = '/content/Tea_Betel_Leaf_Final_Dataset'
train_dir = os.path.join(final_combined_dir, 'train')
test_dir = os.path.join(final_combined_dir, 'test')
val_dir = os.path.join(final_combined_dir, 'valid')

# --- Output path for the combined zipped dataset ---
# *** IMPORTANT:  This is where the combined dataset will be SAVED on your Google Drive ***
output_zip_path = '/content/drive/MyDrive/Tea_Betel_Leaf_Final_Dataset(1).zip'
```

```python
# --- 3. Unzip Datasets ---
with zipfile.ZipFile(tea_zip_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_tea_folder)

with zipfile.ZipFile(betel_zip_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_betel_folder)
```

```python
# --- 4. Create the Combined Dataset Directory Structure ---
os.makedirs(final_combined_dir, exist_ok=True)
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
```

```python
# --- 5. Function to Copy and Rename Classes ---
def copy_and_rename_classes(source_base_dir, dest_base_dir, prefix):
    """Copies and renames class directories from source to destination."""
    for class_name in os.listdir(source_base_dir):
        source_class_dir = os.path.join(source_base_dir, class_name)
        if os.path.isdir(source_class_dir):  # Make sure it's a directory
            new_class_name = f"{prefix}_{class_name}"
            dest_class_dir = os.path.join(dest_base_dir, new_class_name)
            os.makedirs(dest_class_dir, exist_ok=True)

            for filename in os.listdir(source_class_dir):
                source_file = os.path.join(source_class_dir, filename)
                dest_file = os.path.join(dest_class_dir, filename)
                shutil.copy(source_file, dest_file)
```

```python
# --- 6. Copy and Rename Classes from Tea Dataset ---
copy_and_rename_classes(os.path.join(extracted_tea_folder, 'Tea_Leaf_Final_Dataset', 'train'), train_dir, 'tea')
copy_and_rename_classes(os.path.join(extracted_tea_folder, 'Tea_Leaf_Final_Dataset', 'test'), test_dir, 'tea')
copy_and_rename_classes(os.path.join(extracted_tea_folder, 'Tea_Leaf_Final_Dataset', 'valid'), val_dir, 'tea')
```

```python
# --- 7. Copy and Rename Classes from Betel Dataset ---
copy_and_rename_classes(os.path.join(extracted_betel_folder, 'train'), train_dir, 'betel')
copy_and_rename_classes(os.path.join(extracted_betel_folder, 'test'), test_dir, 'betel')
copy_and_rename_classes(os.path.join(extracted_betel_folder, 'val'), val_dir, 'betel')
```

```python
# --- 8. Zip the Final Combined Dataset and Save to Google Drive ---
def zipdir(path, ziph):
    for root, dirs, files in os.walk(path):
        for file in files:
            rel_path = os.path.relpath(os.path.join(root, file), os.path.join(path, '..'))
            ziph.write(os.path.join(root, file), rel_path)
```

```python
with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipdir(final_combined_dir, zipf)  # Zip and save directly to Google Drive
```

```python
# --- 9. Cleanup (Optional) ---
shutil.rmtree(extracted_tea_folder)
shutil.rmtree(extracted_betel_folder)
shutil.rmtree(final_combined_dir) # Clean up the unzipped combined directory in Colab

print(f"Combined dataset created and zipped to: {output_zip_path}")
```
